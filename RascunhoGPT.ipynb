```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Exemplo de DataFrame
df = pd.DataFrame({
    'Nome': ['João', 'Ana (W)', 'Paulo', 'Maria (W)'],
    'Idade': [25, 30, 22, 28],
    'Salário': [2500, 3000, 2200, 3200]
})

# Criar uma figura com 2 subplots (1 linha, 2 colunas)
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Histograma do Salário
sns.histplot(df['Salário'], bins=5, ax=axs[0], kde=True)
axs[0].set_title('Histograma de Salários')
axs[0].set_xlabel('Salário')
axs[0].set_ylabel('Frequência')

# Boxplot da Idade
sns.boxplot(x='Idade', data=df, ax=axs[1])
axs[1].set_title('Boxplot de Idades')
axs[1].set_xlabel('Idade')

# Ajustar o layout
plt.tight_layout()
plt.show()
```
