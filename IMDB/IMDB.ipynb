### 20. Classificar os filmes baseados em avaliação (Excelente, bom e mediano)
O critério a ser atendido
* Excelente > 8.0
* Bom >= 6
* Mediano < 6


```python
def classificacao(nota):
  if nota>=8.0:
    return "Excelente"
  elif nota >= 6.5:
    return "Bom"
  else:
    return "Mediano"
```

```python
df2 = df.copy()
df2.head()
```

```python
df2['classificacao'] = df2['Rating'].apply(classificacao)
df2.head(10)
```

### 22. Listar todos os gêneros de filmes de forma única

```python
df['Genre']
```

```python
# convertendo para uma lista
lista1 = []
for valor in df['Genre']:
  lista1.append(valor.split(','))
```

```python
# separando os itens de um mesmo grupo
one_d = []
for item in lista1:
  for item1 in item:
    one_d.append(item1)

one_d
```

```python
# separando os itens de forma unica
genero = []
for gen in one_d:
  if gen not in genero:
    genero.append(gen)

genero
```


### 23. Encontrar a quantidade de filmes de cada gênero

```python
cont_gen = []
for item in genero:
  x = df['Genre'].str.contains(item).sum()
  cont_gen.append(x)
  print(f'{item}: {x}')
```

```python
sns.barplot(None,x=genero,y=cont_gen)
plt.xticks(rotation=90)
plt.show()
```

```python
ano = np.sort(df['Year'].unique())
ano
```

```python
df[df['Year']==2006]['Genre'].str.contains('Action').sum()
```

```python
matriz_anogenero = np.zeros((len(ano),len(genero)))
n = len(ano)
p = len(genero)
for yy in range(n):
  for gg in range(p):
    matriz_anogenero[yy,gg] = df[df['Year']==ano[yy]]['Genre'].str.contains(genero[gg]).sum()

```

```python
my_df_anogenero = pd.DataFrame(matriz_anogenero,index=ano,columns=genero)
my_df_anogenero
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(my_df_anogenero,annot=True,fmt="1.0f",cmap="crest")
```
