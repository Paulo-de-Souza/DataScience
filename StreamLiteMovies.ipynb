```python
%%writefile testes_stream_sidebar.py
import streamlit as st
from streamlit_option_menu import option_menu

st.set_page_config(layout="wide")

with st.sidebar:
  selected = option_menu(
    menu_title = "Menu",
    options = ["Home", "Projects"],
    menu_icon="cast", default_index=0,
  )

if selected == "Home":
  st.title(f'You have selected {selected}')
if selected == "Projects":
  st.title(f'You have selected {selected}')


```

```python
%%writefile demo.py
import streamlit as st
import pandas as pd
from PIL import Image

st.set_page_config(layout="wide")

path = '/content/drive/MyDrive/Colab Notebooks/new_watchlist_complete.csv'
st.title('Watchlist no StreamLit')
#st.write('Lista de Filmes na Watchlist')
df2 = pd.read_csv(path)
#st.write('Selecione um Filme')

movie = st.selectbox("",df2["Filme"])
info_movie = df2[df2["Filme"]==movie]

index_movie = df2[df2["Filme"]==movie].index.values
index_movie = index_movie[0]

ano_filme = info_movie["Lancamento"][index_movie]
Diretor_filme = info_movie["Diretor"][index_movie]
Generos_filme =	info_movie["Generos"][index_movie]
Atores_filme =	info_movie["Atores"][index_movie]
Avaliacao_filme = info_movie["Avaliacao"][index_movie]
Paises = info_movie["Paises"][index_movie]
Duracao = info_movie["Duracao"][index_movie]

st.divider()
st.subheader(f"Filme: {info_movie['Filme'][index_movie]}")

col1, col2 = st.columns(2)

with col1:
  link_capa = info_movie["link_imagem"][index_movie]
  st.image(link_capa)

with col2:
  st.write(f"Lancamento: {int(ano_filme)}")
  st.write(f"Diretor: {Diretor_filme}")
  st.write(f"Genero: {Generos_filme}")
  st.write(f"Avaliaco Media: {Avaliacao_filme}")
  st.write(f"Duracao: {Duracao}")
  st.write(f"Elenco: {Atores_filme}")

```
